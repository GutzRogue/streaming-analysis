# Analyse des abonnés - Devoir Hadoop/Spark

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import sum, avg, count, col, round

spark = SparkSession.builder.appName('Analyse Abonnes').getOrCreate()
```

```python
df = spark.read.csv(
    "hdfs://namenode:9000/data/abonnes_cluster_demo.csv",
    header=True,
    inferSchema=True
)
df.show(5)
df.printSchema()
```

## 1. Nombre total d'abonnés

```python
df.count()
```

## 2. Revenu total

```python
df.agg(round(sum('montant'),2).alias('revenu_total')).show()
```

## 3. Revenu par type d'abonnement

```python
df.groupBy('abonnement').agg(round(sum('montant'),2).alias('revenu_total')).orderBy(col('revenu_total').desc()).show()
```

## 4. Âge moyen

```python
df.agg(round(avg('age'),2).alias('age_moyen')).show()
```

## 5. Nombre d'abonnés par ville

```python
df.groupBy('ville').agg(count('*').alias('nombre_abonnes')).orderBy(col('nombre_abonnes').desc()).show()
```

## 6. Ville la plus représentée

```python
df.groupBy('ville').agg(count('*').alias('nombre_abonnes')).orderBy(col('nombre_abonnes').desc()).limit(1).show()
```

## 7. Durée moyenne d'utilisation

```python
df.agg(round(avg('duree_utilisation_mois'),2).alias('duree_moyenne_mois')).show()
```
